```python
import requests
```

```python
URL = 'https://directory.goodonyou.eco/categories/fashion'

headers = {
    "User-Agent": "MyPythonScript/1.0 (contact@example.com)"
}

response = requests.get(URL, headers=headers)
```

```python
from bs4 import BeautifulSoup as BS
```

```python
soup = BS(response.text)
```

COMPANY NAME

```python
companies = [
    s.text.strip()
    for s in soup.find_all('span', class_='LinkNext__LinkBody-sc-hdzulb-1 gfkCxl')
    if s.text.strip()
]
companies
```

RATINGS

```python
ratings = [
    s.text.strip()
    for s in soup.find_all('span', class_='StyledText-sc-1sadyjn-0 itVbFZ')
    if s.text.strip()
]
ratings
```

```python
price_list = []

price_divs = soup.find_all('div', class_='PriceRating__InlineBox-sc-cbcbni-0')

for div in price_divs:
    spans = div.find_all('span')
    price = 0
    for s in spans:
        style = s.get("style", "").replace(" ", "").lower()
        if "opacity:1" in style:
            price += 1
    price_list.append(price)
```

```python
price_list
```

```python
cards = soup.find_all('div', class_='StyledBox-sc-13pk1d4-0 kXaaht')

ratings = []
countries = []
prices = []

for card in cards:
    inner = card.find('div', class_='StyledBox-sc-13pk1d4-0 kgWyRM')

    # --- Rating & Country ---
    spans = inner.find_all('span', class_='StyledText-sc-1sadyjn-0 itVbFZ')
    rating = spans[0].get_text(strip=True)
    country = spans[1].get_text(strip=True)

    # --- Price ---
    price_div = inner.find('div', class_='PriceRating__InlineBox-sc-cbcbni-0')
    price_spans = price_div.find_all('span')

    # Count ONLY the dark "$" signs (opacity = 1)
    price_count = 0
    for s in price_spans:
        style = s.get("style", "")
        if "opacity: 1" in style:
            price_count += 1

    # store results
    ratings.append(rating)
    countries.append(country)
    prices.append(price_count)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


price_dist = pd.crosstab(
    df_all["rating"],
    df_all["price_level"],
    normalize="index"
) * 100


price_dist = price_dist[[1, 2, 3, 4]]

plt.figure(figsize=(12, 7))

bottom = np.zeros(len(price_dist))
colors = ["#68b0ab", "#8fc0a9", "#efa5a5", "#cb9070"]  # color palette (1–4)


for i, col in enumerate(price_dist.columns):
    bars = plt.bar(
        price_dist.index,
        price_dist[col],
        bottom=bottom,
        label=f"Price Level {col}",
        color=colors[i],
        edgecolor="white",  
        linewidth=1
    )

  
    for j, value in enumerate(price_dist[col]):
        if value < 5:
       
            y_pos = bottom[j] + value + 1
            va = "bottom"
        else:
        
            y_pos = bottom[j] + value / 2
            va = "center"

        plt.text(
            j, y_pos, f"{value:.1f}%",
            ha='center', va=va,
            fontsize=9, color='black'
        )

    # Update bottom for next stack
    bottom += price_dist[col]


plt.title("Price Level Composition by Sustainability Rating", fontsize=16, pad=15)
plt.ylabel("Percentage (%)", fontsize=12)
plt.xlabel("Rating", fontsize=12)
plt.xticks(rotation=20)
plt.legend(title="Price Levels", loc="upper left")
plt.ylim(0, 105)
plt.tight_layout()
plt.show()
```

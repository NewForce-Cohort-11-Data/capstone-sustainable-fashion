```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sf_ff_5.csv")
```

```python
df
```

```python
df.columns = ["Week", "Sustainable_Fashion", "Fast_Fashion"]
```

```python
df
```

```python
df.dtypes
```

```python
df["Week"] = pd.to_datetime(df["Week"], errors="coerce")
```

```python
plt.figure(figsize=(12,6))

plt.plot(df["Week"], df["Sustainable_Fashion"], label="Sustainable Fashion", linewidth=2)
plt.plot(df["Week"], df["Fast_Fashion"], label="Fast Fashion", linewidth=2)

plt.title("Sustainable vs Fast Fashion (Past 5 Years)")
plt.xlabel("Year")
plt.ylabel("Search Interest")
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()

plt.show()
```
